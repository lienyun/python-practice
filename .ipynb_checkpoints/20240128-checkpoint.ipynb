```python
import pandas as pd
import numpy as np

np.random.seed(100)
df = pd.DataFrame({
    'date':pd.date_range('2020-1-1', periods = 10, freq = 'D'),
    'y1':np.random.randint(50, 151, size = 10),
    'y2':np.random.randint(60, 201, size = 10),
})
df
```

```python
import matplotlib.pyplot as plt
import datetime as dt
plt.rcParams['font.family'] = 'Microsoft Yahei'
ax = df['y1'].plot(#x = 'date',
        #y = 'y1', 
        figsize = (5, 3),  #
        marker = 'o',      #
        color = 'red',     #
        xlabel = '',   #X
        ylabel = '', #Y
        title = '', #
        #xlim = (dt.datetime(2020, 1, 1), dt.datetime(2020, 1, 15)),#X
        ylim = (50, 200),   #Y\
        markerfacecolor = 'purple', #marker(mfc='purple')
        markersize = 14, #marker (ms=14)
        markeredgecolor = 'yellow' #marker border color(mec)
        #kind = 'bar'
       )

df['y2'].plot(#x = 'date',
        #y = 'y2',
        figsize = (5, 3),
        #marker = '^',      #
        color = 'green',
        ylim = (50, 200),   #Y
        secondary_y = True,
        ax = ax)

datalist = list(map(lambda x: dt.datetime.strftime(x, "%Y/%m/%d"),df['date']))
plt.xticks(range(len(df.index)),datalist, rotation = 30)
plt.ylabel('\n\n', rotation = 0, labelpad = 20) #y

#
#plt.savefig('.png',format = 'png') #
plt.savefig('.png',format = 'png',bbox_inches = 'tight') #
```

```python
#
import matplotlib.pyplot as plt

np.random.seed(100)
df = pd.DataFrame({
    'date':pd.date_range('2020-1-1', periods = 10, freq = 'D'),
    'y1':np.random.randint(50, 151, size = 10),
    'y2':np.random.randint(60, 201, size = 10),
})
#plt.plot(x = df['date'], y = df['y1']) 
plt.figure(figsize = (5,3),
           facecolor = 'lightblue',
           edgecolor = 'g',
           linewidth = 3
          )
# marker
axes = plt.axes()
axes.set_facecolor = '#159A9C'


plt.plot(df['date'],#xy
         df['y1'],
         c = 'r', #c=color, r=red, green=g, blue=b, black=k
         lw = 3, #linewidth
         ls = '--', #linestyle
         marker = 'p', #marker
         ms = 14,
         mfc = 'pink',
         mec = 'b'
        ) 

#
plt.bar(df['date'],
       height = df['y2'])

plt.xticks(rotation = 30)
plt.xlabel('')
plt.ylabel('\n\n',rotation = 0, labelpad = 20)
plt.title('')
plt.legend([''], loc = 9, facecolor = '#B4BEC9', edgecolor = '#DEEFE7')#legend[] loc0-9
plt.ylim(50,250) # yrange
plt.grid(True, color = '#eeeeee',) #

#marker
for x,y in zip(df['date'],df['y1']): #zip: colzip
    text = '{}'.format(y)
    plt.text(x,y+10,text,color = '#04BF9D', ha = 'center') #y+10

plt.savefig('2.png',format = 'png',bbox_inches = 'tight')
```

```python
import matplotlib.pyplot as plt

np.random.seed(100)
df = pd.DataFrame({
    'date':pd.date_range('2020-1-1', periods = 10, freq = 'D'),
    'y1':np.random.randint(50, 151, size = 10),
    'y2':np.random.randint(60, 201, size = 10),
})

plt.figure(figsize = (10,10),
          facecolor = '#DEEFE7')
plt.subplot(321)#(row,col,index)
plt.plot(df['y1'])
plt.xticks(range(0,len(df.index),2),
           list(map(lambda x: dt.datetime.strftime(x, "%Y/%m/%d"),df['date'][::2])),
           rotation = 30)

#2
plt.subplot(322)
plt.bar(df.index, height = df['y1'])
plt.xticks(range(0,len(df.index),2),
           list(map(lambda x: dt.datetime.strftime(x, "%Y/%m/%d"),df['date'][::2])),
           rotation = 30)

plt.subplot(323)
plt.barh(df.index, width = df['y1'])
plt.yticks(range(0,len(df.index),2),
           list(map(lambda x: dt.datetime.strftime(x, "%Y/%m/%d"),df['date'][::2])),
           rotation = 30)

plt.subplot(324)
plt.scatter(df.index, df['y2'])
plt.xticks(range(0,len(df.index),2),
           list(map(lambda x: dt.datetime.strftime(x, "%Y/%m/%d"),df['date'][::2])),
           rotation = 30)


#
plt.subplot(325)
width = 0.3 #bar
plt.bar(df.index, width = width, height = df['y1'])
plt.bar(df.index+width, width = width, height = df['y2'])
plt.xticks(range(0,len(df.index),2),
           list(map(lambda x: dt.datetime.strftime(x, "%Y/%m/%d"),df['date'][::2])),
           rotation = 30)

#
plt.subplot(326)
plt.bar(df.index, height = df['y1'])
plt.bar(df.index, height = df['y2'],bottom = df['y1'].tolist())




plt.suptitle('',fontsize = 18)
#X hspace
plt.subplots_adjust(hspace = 0.3)
```

```python
import matplotlib.pyplot as plt

np.random.seed(100)
df = pd.DataFrame({
    'date':pd.date_range('2020-1-1', periods = 10, freq = 'D'),
    'y1':np.random.randint(50, 151, size = 10),
    'y2':np.random.randint(60, 201, size = 10),
})

#: !
fig, ax = plt.subplots(3,2,figsize = (10,10))
#axmatlab

#
ax[0,0].plot(df.index, df['y1'])
ax[0,0].set_xticks(range(0,len(df.index),2),
           list(map(lambda x: dt.datetime.strftime(x, "%Y/%m/%d"),df['date'][::2])),
           rotation = 30)
ax[0,0].text(4,110, '(0,0)')

ax[0,1].bar(df.index, height = df['y1'])
ax[0,1].set_xticks(range(0,len(df.index),2),
           list(map(lambda x: dt.datetime.strftime(x, "%Y/%m/%d"),df['date'][::2])),
           rotation = 30)

#[0,0]
ax[0,0].set_ylabel('\n\n',rotation = 0, labelpad = 20)



ax[1,0].barh(df.index, width = df['y1'])
ax[1,0].set_yticks(range(0,len(df.index),2),
           list(map(lambda x: dt.datetime.strftime(x, "%Y/%m/%d"),df['date'][::2])),
           rotation = 30)
ax[1,0].text(110,5, 'barh(1,0)',fontsize = 16, color = 'red')


ax[1,1].scatter(df.index, df['y2'])
ax[1,1].set_xticks(range(0,len(df.index),2),
           list(map(lambda x: dt.datetime.strftime(x, "%Y/%m/%d"),df['date'][::2])),
           rotation = 30)
ax[1,1].text(4,110, 'scatter(1,1)',fontsize = 16, color = 'red')

width = 0.3 #bar
ax[2,0].bar(df.index, width = width, height = df['y1'])
ax[2,0].bar(df.index+width, width = width, height = df['y2'])
ax[2,0].set_xticks(range(0,len(df.index),2),
           list(map(lambda x: dt.datetime.strftime(x, "%Y/%m/%d"),df['date'][::2])),
           rotation = 30)

ax[2][1].bar(df.index, height = df['y1'])
ax[2][1].bar(df.index, height = df['y2'],bottom = df['y1'].tolist())

#X hspace
plt.subplots_adjust(hspace = 0.3)
```
